WHAT DOES ensue annotations DO ?

```python

def get_product(x:int,y:int) -> int:
  return x * y
```

```python
print(get_product(2,3))
```

```python
print(get_product(23,"23"))
```

```python
# to prevent the above bugs where it ensures these
# kinds of bugs like type conversion or not returning
# the proper type 
from ensure import ensure_annotations
```

```python
@ensure_annotations
def get_product(x:int,y:int) -> int:
  return x * y
```

```python
print(get_product(2,3))
```

```python
print(get_product(23,"23"))
```
